```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import scipy.sparse.linalg as sp_linalg
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.neighbors import kneighbors_graph

from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
def aff_matrix(S,sigma):
    n = S.shape #n taille de s n*n
    A = np.zeros((n[0],n[0])) #creation d'une matrice de taille n*n
    for i in range(n[0]):
        for j in range(n[0]):
            A[i][j] = np.exp(-(np.linalg.norm(S[i]-S[j])**2)/(2*(sigma**2)))
    return A

def d_construction(A):
    D = np.zeros(A.shape)
    for i in range(A.shape[0]):
        D[i][i] = A[i][0]
        for j in range(1,A.shape[1]):
            D[i][i] = D[i][i] + A[i][j]
        D[i][i] = D[i][i]**(-0.5)
    return D.dot(A).dot(D)

def gram_schmidt_columns(X):
    Q, R = np.linalg.qr(X)
    return Q

def cluster(L,k):
    valPropre, vectPropre = sp_linalg.eigs(L,k)
    X = gram_schmidt_columns(vectPropre.real)
    Y = np.zeros(X.shape)
    norme = np.linalg.norm(X, axis=1, ord=None)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Y[i][j] = X[i][j] / norme[i]
    kmeans = KMeans(n_clusters=k, random_state=1231)
    return kmeans.fit(X).labels_

def gen_kernels(filename, k, m, sigma):
    q = aff_matrix(m,sigma)
    q = d_construction(q)
    labels = cluster(q,k)
    plt.figure()
    plt.scatter(m[:,0], m[:,1], c=labels, s=50, cmap='viridis')
    plt.title(filename + ", Kernels method, k=" + str(k) + ", sigma=" + str(sigma))
    plt.savefig(filename + '/' + filename + '_kernels__k=' + str(k) + '_sigma=' + str(sigma) + '_.png')
    
def gen_dendogram(filename, algoname, m, k):
    connectivity = kneighbors_graph(m, n_neighbors=k, include_self=False)
    connectivity = 0.5 * (connectivity + connectivity.T)
    labels = AgglomerativeClustering(n_clusters=k, linkage=algoname, connectivity=connectivity).fit(m).labels_
    plt.figure()
    plt.title(filename + ", " + algoname + " dendrogram, k=" + str(k))
    plt.scatter(m[:,0], m[:,1], c=labels, s=50, cmap='viridis')
    plt.savefig(filename + '/' + filename + '_' + algoname + '_dedrogram__k=' + str(k) + '_.png')
    
def gen_dbscan(filename, m, eps, min_samples):
    labels = DBSCAN(eps=eps, min_samples=min_samples).fit(m).labels_
    plt.figure()
    plt.title(filename + ", DBSCAN, min_samples=" + str(min_samples) + ", eps=" + str(eps))
    plt.scatter(m[:,0], m[:,1], c=labels, s=50, cmap='viridis')
    plt.savefig(filename + '/' + filename + '_dbscan__min_samples=' + str(min_samples) + '_eps=' + str(eps) + '_.png')
    
def gen_kmeans(filename, m, k):
    kmeans = KMeans(n_clusters=k, random_state=1231)
    labels = kmeans.fit(m).labels_
    plt.figure()
    plt.title(filename + ", KMeans, k=" + str(k))
    plt.scatter(m[:,0], m[:,1], c=labels, s=50, cmap='viridis')
    plt.savefig(filename + '/' + filename + '_kmeans__k=' + str(k) + '_.png')
    
def read_dataset_2d(filename):
    with open(filename + ".txt") as fileIn:
        n1, n2 = (float(s) for s in fileIn.readline().split())
        m = [n1,n2]
        for line in fileIn:
            n1, n2 = (float(s) for s in line.split())
            v = [n1,n2]
            m = np.c_[m,v]
    return m.T
    
def read_dataset_3d(filename):
    with open(filename + ".txt") as fileIn:
        n1, n2, n3 = (float(s) for s in fileIn.readline().split())
        m = [n1,n2]
        print(m)
        for line in fileIn:
            n1, n2, n3 = (float(s) for s in line.split())
            v = [n1,n2]
            m = np.c_[m,v]
    return m.T   
```

```python
filename = "jain"
k = 2
m = read_dataset_3d(filename)

# Kernel method
#gen_kernels(filename, k, m, 0.25) # Data=Compound+jain : ArpackNoConvergence: ARPACK error -1: No convergence (3991 iterations, 0/9 eigenvectors converged) [ARPACK error -14: DNAUPD  did not find any eigenvalues to sufficient accuracy.]
#gen_kernels(filename, k, m, 0.5) # Data=jain : ArpackNoConvergence: ARPACK error -1: No convergence (3731 iterations, 0/9 eigenvectors converged) [ARPACK error -14: DNAUPD  did not find any eigenvalues to sufficient accuracy.]
gen_kernels(filename, k, m, 0.675)
gen_kernels(filename, k, m, 1)
gen_kernels(filename, k, m, 5)

# Dendogram
gen_dendogram(filename, 'ward', m, k)
gen_dendogram(filename, 'complete', m, k)
gen_dendogram(filename, 'average', m, k)

# DBSCAN
gen_dbscan(filename, m, 4, 6)
gen_dbscan(filename, m, 3, 6)
gen_dbscan(filename, m, 3, 22)

# KMEANS
gen_kmeans(filename, m, k)
gen_kmeans(filename, m, k)

plt.show()
```
